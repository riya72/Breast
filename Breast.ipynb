# Spliting Out Dataset For Training and Testing by using some particular data

```python
breast_y_label=breast["diagnosis1"]  #dependent variable
breast_x_labels=breast[["radius_mean","perimeter_mean","area_mean"]] #independant variable
```

# Feature Scaling

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(breast_x_labels, breast_y_label, test_size=0.2, random_state=42)
```

# Training and Testing the Model 

#  1)Logistic Regrassion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1=LogisticRegression()
model1.fit(x_train,y_train)
model1_prediction=model1.predict(x_test)
print("Accuracy of Logistic Regression:",accuracy_score(y_test,model1_prediction)) #match the y_test value with model1_prediction value and find out how much data is accurate
```

# 2)Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model2=DecisionTreeClassifier()
model2.fit(x_train,y_train)
model2_prediction=model1.predict(x_test)
print("Accuracy of DecisionTreeClassifier:",accuracy_score(y_test,model2_prediction))
```

# 3)Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score
model3=DecisionTreeClassifier()
model3.fit(x_train,y_train)
model3_prediction=model1.predict(x_test)
print("Accuracy of RandomForestClassifier :",accuracy_score(y_test,model3_prediction))
```

#  Spliting Out Dataset For Training and Testing by using all data

```python
breast_y1_label=breast["diagnosis1"]  #dependent variable
breast_x1_labels=breast.drop(('diagnosis1'),axis=1)
```

# Training and Testing the Model

# 1)Logistic Regrassion

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1=LogisticRegression()
model1.fit(x1_train,y1_train)
model1_prediction=model1.predict(x1_test)
print("Accuracy of Logistic Regression:",accuracy_score(y1_test,model1_prediction))
```

# 2)Decission Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model2=DecisionTreeClassifier()
model2.fit(x1_train,y1_train)
model2_prediction=model1.predict(x1_test)
print("Accuracy of DecisionTreeClassifier :",accuracy_score(y1_test,model2_prediction))
```
