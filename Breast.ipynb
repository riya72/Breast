#  Converting the Catgorical value into Numerical Value

```python
Breast['diagnosis'].unique()
```

```python
# one hot encoding
diagnosis1=pd.get_dummies(Breast["diagnosis"],drop_first=True)
diagnosis1
```

```python
Breast=pd.concat([Breast,diagnosis1],axis=1)
```

```python
Breast.head()
```

```python
Breast.drop(["diagnosis"],axis=1,inplace=True)
```

```python
Breast.head()
```
